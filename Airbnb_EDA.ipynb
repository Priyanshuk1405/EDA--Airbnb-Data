```python
# Converting list into array to get the data for initial operations like finding Mean, Median and Standard deviation

data = airbnb[['price','minimum nights','number of reviews','availability 365']]
arr = data.to_numpy()
print(arr)

# data = np.array(df[['price','minimum nights','number of reviews','availability 365']])
```

```python
#Task 3 find mean median and Std of each column
# For finding mathematics operation we need to remove $ sign to get a perfect float value because of string interpretation we can't able to find such things.
price = arr[:,0]
# print(price)
price_float = [] # created a empty list for append that prices in diffrent format.
for i in price:
  i = str(i)
  price_1 = i.replace('$','').replace(',','').strip()
  price_float.append(float(price_1))

print(price_float)
```

```python
#Filter data in various condition, where prices is less than 250, availability 365 is greater than 150 and minimum nights > 5
price_series = pd.Series(price_float)
filtered_data = airbnb[(price_series < 250) & (airbnb['availability 365'].astype(float) > 150) & (airbnb['minimum nights'].astype(float) > 5)]
display(filtered_data.head(10))
```

```python
airbnb.isnull().sum()
```

```python
# i want to add that float price columns in main dataframe
airbnb['price_float'] = price_float
```

```python
airbnb['minimum nights'] = pd.to_numeric(airbnb['minimum nights'], errors='coerce')
airbnb['number of reviews'] = pd.to_numeric(airbnb['number of reviews'], errors='coerce')
airbnb['availability 365'] = pd.to_numeric(airbnb['availability 365'], errors='coerce')

# Now i need to filter the data with adn remove null values and fill with 0.00 in price, minimum heights, availability 365, number of reviews.

columns_to_fill = ['price','minimum nights','number of reviews','availability 365']

airbnb[columns_to_fill].fillna(0.00)
airbnb['price_float'] = price_float
# print(airbnb.head())
mean_of_airbnb = airbnb[['price_float', 'minimum nights','number of reviews','availability 365']].mean()
print(mean_of_airbnb)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
airbnb.head()
```

```python
plt.Figure(figsize=(10,6))
sns.boxplot( y= airbnb['price_float'])
plt.title("Distribution of Prices", fontstyle = "italic")
plt.ylim(300,1000) # To ignore the extreme outliers
plt.ylabel("Price in ($)")
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.countplot(data = airbnb, x = "room type", hue ='room type')
plt.title("Room type distribution")
plt.xlabel("Room Type")
plt.ylabel("Total number of customer")
plt.show()
```

```python
# top 10 neighbourhood by listing count
top_neighbour = airbnb['neighbourhood'].value_counts().nlargest(10)
# print(top_neighbour)
plt.figure(figsize = (10,6))
sns.barplot(x = top_neighbour.values, y = top_neighbour.index, color="teal")
plt.title("Top 10 neighbourhood by listing count")
plt.xlabel("No. of customer")
plt.ylabel("Neighbourhoods")
plt.show()
```

```python
plt.figure(figsize= (10,6))
sns.boxplot(x = "room type", y = "price_float", data=airbnb, hue="room type")
plt.title("Price affected by Room type")
plt.ylabel("Prices")
plt.xlabel("Room Type")
plt.show()
```

```python

filtered = airbnb[(airbnb['neighbourhood'].isin(top_neighbour.index)) & (airbnb['price_float'] < 1000)]

plt.figure(figsize=(6,10))
sns.violinplot(data=filtered, y='neighbourhood', x='price_float', palette='tab10')
plt.title('Violin Plot of Price by Top 10 Neighborhoods (Under $500)')
plt.ylabel('Neighborhood')
plt.xlabel('Price ($)')
plt.show()



```

```python

plt.figure(figsize = (10,6))
sns.histplot( airbnb['availability 365'], bins = 50, kde = True)
plt.title("Availability over the Year")
plt.xlabel("Availability")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data = airbnb[airbnb['price_float']<1000], x = 'availability 365', y = 'price_float')
plt.title('Price vs Availability')
plt.xlabel('Availability (days/year)')
plt.ylabel('Price ($)')
plt.show()
```
